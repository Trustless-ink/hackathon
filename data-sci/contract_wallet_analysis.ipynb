# Contract wallet analysis

This notebook analyzes the wallets for evolved apes and frostie contract, both of which ended up rug-pulling. We want to see the action of those wallets.


```python
import numpy as np
import pandas as pd
import plotly.express as px
```

# evolved apes

```python
fname = '/home/zhiwei/Documents/trustless/evolved_aps_wallet1.csv'
fname0 = '/home/zhiwei/Documents/trustless/evolved_aps_wallet2.csv'

df0 =pd.concat([pd.read_csv(fname,index_col=False),pd.read_csv(fname0,index_col=False)],axis=0,ignore_index=True) 
cols = ['DateTime','Value_IN(ETH)','Value_OUT(ETH)']

fname1 = '/home/zhiwei/Documents/trustless/evolved_aps_wallet1_internal.csv'
fname2 = '/home/zhiwei/Documents/trustless/evolved_aps_wallet2_internal.csv'
df_int =pd.concat([pd.read_csv(fname1,index_col=False),pd.read_csv(fname2,index_col=False)],axis=0,ignore_index=True) 
df = pd.concat([df0,df_int],ignore_index=True).sort_values(by='DateTime')
wallet_total = df['Value_OUT(ETH)'].sum()
wallet_removal =  df['Value_OUT(ETH)'].cumsum()
df['wallet_removal']=df['Value_OUT(ETH)'].cumsum()
df['wallet_remains']=df['Value_IN(ETH)'].cumsum()-df['Value_OUT(ETH)'].cumsum()

f_ape=px.line(df,x='DateTime',y='wallet_remains')

df_ape = df.copy()
```

# Frostie

```python
fname = '/home/zhiwei/Documents/trustless/frostie_contract_wallet.csv'
cols = ['DateTime','Value_IN(ETH)','Value_OUT(ETH)']
df0 = pd.read_csv(fname,index_col=False)
df_int = pd.read_csv('frostie_contract_wallet_internal.csv',index_col=False)
df = pd.concat([df0[cols],df_int[cols]],axis=0,ignore_index=True).sort_values(by='DateTime')

wallet_add = df['Value_IN(ETH)'].cumsum()
wallet_removal =  df['Value_OUT(ETH)'].cumsum()
df['wallet_remains']=wallet_add-wallet_removal

f_frostie = px.line(df,x='DateTime',y='wallet_remains',title='frostie')
df_frostie = df.copy()
```

# compare the two rug-pull projects -- visualized

```python
dfs = [df_ape,df_frostie]
labels = ['evolved apes','frostie']
from turtle import width
from plotly.subplots import make_subplots
import plotly.graph_objects as go


fig = make_subplots(rows=1, cols=len(dfs),subplot_titles=labels) 
for k,df in enumerate(dfs):
    fig.append_trace(go.Scatter(
        x=df['DateTime'],
        y=df['wallet_remains'],
    ), row=1, col=k+1)
fig.update_layout(
    width=900,
    height=400,)
fig.show()
```
